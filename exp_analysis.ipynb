## Bell Curves

```python
# Extract the columns as separate NumPy arrays
columns_to_plot = ['advantage_mean', 'believability_mean', 'distinctiveness_mean', 'purchase_likelihood_mean', 'premiumness_mean']
bell_curve_colors = ['red', 'blue', 'green', 'purple', 'orange']  # Define colors for each bell curve

plt.figure(figsize=(8, 6))

for col, color in zip(columns_to_plot, bell_curve_colors):
    # Estimate parameters (mean and standard deviation) for each column
    mean = np.mean(slice[col])
    std_dev = np.std(slice[col])
    
    # Generate data points for the bell curve
    x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 1000)
    y = norm.pdf(x, mean, std_dev)
    
    # Plot the bell curve for the current column
    plt.plot(x, y, label=col, color=color)

# Add labels and title
plt.xlabel('Mean')
plt.ylabel('Probability Density')
plt.title('Bell Curves for Different Columns')
plt.legend()
plt.grid(True)
plt.show()
```

## Chi-Squared Test:

Seeing if presence of "taste profile_mapped_clean_Cheesy" and "purchase_likelihood_mean" have a relationship

```python
cheesy_purchaselikelihood_table = pd.crosstab(slice['taste profile_mapped_clean_Cheesy'], slice['purchase_likelihood_mean'])
# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(cheesy_purchaselikelihood_table)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

Based on the p-value of 0.4824, there is not enough evidence to conclude that there is a statistically significant association between the presence of a specific word in an ad campaign ("taste profile_mapped_clean_Cheesy") and the likelihood of purchase ("purchase_likelihood_mean").
The high p-value suggests that any observed differences between the groups are likely due to chance, and there is no strong indication that the presence of the word is related to purchase likelihood in a significant way.

## K-Means Clustering

```python
# DataFrame of selected_columns
df_selected_columns = slice[selected_columns]
# Use the Elbow Method to find the optimal number of clusters
distortions = []
K = range(1, 11)  # Test different values of k
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init='auto')
    kmeans.fit(df_selected_columns)
    distortions.append(kmeans.inertia_)

# Plot the Elbow graph
plt.figure(figsize=(8, 6))
plt.plot(K, distortions, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Distortion')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Choose the optimal k based on the Elbow Method
optimal_k = 2

# Fit K-Means clustering with the selected k
kmeans = KMeans(n_clusters=optimal_k, random_state=0, n_init='auto')
cluster_labels = kmeans.fit_predict(df_selected_columns)

# Add the cluster labels to your DataFrame
df_selected_columns.loc[:, 'Cluster'] = cluster_labels
```

```python
# Apply PCA to reduce the data to 2 dimensions
pca = PCA(n_components=5)
reduced_data = pca.fit_transform(df_selected_columns)

# Create a scatter plot of the reduced data color-coded by cluster
plt.figure(figsize=(8, 6))

# Scatter plot using the reduced data (2D) and color-coded by cluster
scatter = plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=cluster_labels, cmap='rainbow')

# Customize the plot
plt.title('K-Means Clustering (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster')
# Get the loadings matrix
loadings = pca.components_

# Show the plot
plt.show()

# Show components
loadings
```
