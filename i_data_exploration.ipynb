```python
# Define the keywords you want to search for
keywords = ['concept_id', 'ingredients mentioned in the flavor names', 'distinctiveness_mean']

# Filter columns that contain any of the keywords
flavor_filter = [col for col in slice.columns if any(keyword in col for keyword in keywords)]

# Create a new dataframe with only the filtered columns
flavors_slice = slice[flavor_filter]

# Rename columns dynamically
flavors_slice.columns = flavors_slice.columns.str.replace('ingredients mentioned in the flavor names_mapped_clean_', '')

flavors_slice
```

```python
from scipy import stats

# Assuming you have a DataFrame 'flavors_slice' with the relevant data

# Separate premiumness_mean for Cheese and Chicken
distinctiveness_beef = flavors_slice[flavors_slice['Beef'] == 1]['distinctiveness_mean']
distinctiveness_jalapeno = flavors_slice[flavors_slice['Jalapeno'] == 1]['distinctiveness_mean']

# Perform a t-test to compare premiumness between Cheese and Chicken
t_stat, p_value = stats.ttest_ind(distinctiveness_beef, distinctiveness_jalapeno)

# Output the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Check if the difference is statistically significant at a significance level of 0.05
if p_value < 0.05:
    print("There is a statistically significant difference between Beef and Jalapeno distinctiveness.")
else:
    print("There is no statistically significant difference between Beef and Jalapeno distinctiveness.")
```

```python
import matplotlib.pyplot as plt

# Create separate box plots for Beef = 0 and Beef = 1 using Pandas
flavors_slice.boxplot(column='distinctiveness_mean', by='Beef', vert=False)
plt.title('Box Plot of Distinctiveness Mean for Beef')
plt.xlabel('Beef')
plt.ylabel('Distinctiveness Mean')
plt.suptitle('')  # Remove the default title created by Pandas
plt.show()

# Create separate box plots for Chicken = 0 and Chicken = 1 using Pandas
flavors_slice.boxplot(column='distinctiveness_mean', by='Jalapeno', vert=False)
plt.title('Box Plot of Distinctiveness Mean for Jalapeno')
plt.xlabel('Chicken')
plt.ylabel('Distinctiveness Mean')
plt.suptitle('')  # Remove the default title created by Pandas
plt.show()
```
